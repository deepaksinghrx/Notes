```python
# ============================================================================
class PipelineConfig:
    """
    Centralized configuration class for the entire pipeline.
    All paths, parameters, and constants are defined here.
    """

    def __init__(self, use_google_drive: bool = True):
        """
        Initialize pipeline configuration.

        Args:
            use_google_drive: Whether to use Google Drive for storage
        """
        self.use_google_drive = use_google_drive

        # Google Drive mount point (Colab specific)
        self.drive_mount_point = '/content/drive'
        self.drive_root = f'{self.drive_mount_point}/MyDrive/Mental Health'

        # Dataset location (where CSV files are stored)
        self.dataset_path = f'{self.drive_mount_point}/MyDrive/ColabNotebooks'

        # Local fallback directory
        self.local_root = './outputs'

        # Determine base path
        if use_google_drive:
            self.base_path = self.drive_root
        else:
            self.base_path = self.local_root
            self.dataset_path = './'  # Local fallback for datasets

        # Create directory structure
        self.dirs = {
            'data_raw': f'{self.base_path}/data/raw',
            'data_processed': f'{self.base_path}/data/processed',
            'data_features': f'{self.base_path}/data/features',
            'models': f'{self.base_path}/models',
            'models_tflite': f'{self.base_path}/models/tflite',
            'plots': f'{self.base_path}/plots',
            'plots_eda': f'{self.base_path}/plots/eda',
            'plots_models': f'{self.base_path}/plots/models',
            'plots_explainability': f'{self.base_path}/plots/explainability',
            'reports': f'{self.base_path}/reports',
            'logs': f'{self.base_path}/logs',
            'cache': f'{self.base_path}/cache'
        }

        # Dataset configuration
        self.datasets = {
            'anxiety': {
                'filename': f'{self.dataset_path}/mental_heath_unbanlanced for anxiety.csv',
                'target': 'status',
                'text_column': 'text',
                'drop_cols': ['Unnamed: 0'],
                'type': 'text',
                'task': 'binary_classification',
                'description': 'Anxiety detection from text statements'
            },
            'stress': {
                'filename': f'{self.dataset_path}/Twitter_Full.csv',
                'target': 'labels',
                'text_column': 'text',
                'drop_cols': ['hashtags', 'Unnamed: 3', 'Unnamed: 4'],
                'type': 'text',
                'task': 'binary_classification',
                'description': 'Stress level prediction from text content',
                'sep': ';'
            },
            'depression': {
                'filename': f'{self.dataset_path}/depression_dataset_reddit_cleaned.csv',
                'target': 'is_depression',
                'text_column': 'clean_text',
                'drop_cols': [],
                'type': 'text',
                'task': 'binary_classification',
                'description': 'Depression detection from Reddit posts'
            },
            'bipolar': {
                'filename': f'{self.dataset_path}/data_to_be_cleansed.csv',
                'target': 'target',
                'text_column': 'combined_text_title', # New combined text column
                'drop_cols': ['Unnamed: 0'], # 'text' and 'title' will be dropped in DataCleaner after combination
                'type': 'text',
                'task': 'binary_classification',
                'description': 'Bipolar disorder prediction from combined text (title + text)'
            },
            'suicide': {
                'filename': f'{self.dataset_path}/Suicide_Detection.csv',
                'target': 'class',
                'text_column': 'text',
                'drop_cols': ['Unnamed: 0'],
                'type': 'text',
                'task': 'binary_classification',
                'description': 'Suicide risk detection from text'
            }
        }

        # Model hyperparameters
        self.model_params = {
            'random_forest': {
                'n_estimators': [100, 200],
                'max_depth': [10, 20, None],
                'min_samples_split': [2, 5]
            },
            'xgboost': {
                'n_estimators': [100, 200],
                'max_depth': [5, 10],
                'learning_rate': [0.01, 0.1]
            },
            'logistic_regression': {
                'C': [0.1, 1.0, 10.0],
                'penalty': ['l2'],
                'solver': ['lbfgs']
            }
        }

        # NLP parameters
        self.nlp_params = {
            'max_features_tfidf': 200,
            'ngram_range': (1, 2),
            'min_df': 2,
            'max_df': 0.95,
            'word2vec_dim': 100,
            'word2vec_window': 5,
            'word2vec_min_count': 2
        }

        # Feature engineering parameters
        self.feature_params = {
            'pca_variance': 0.95,
            'n_clusters_kmeans': 5,
            'anomaly_contamination': 0.1
        }

        # Training parameters
        self.training_params = {
            'test_size': 0.2,
            'val_size': 0.1,
            'cv_folds': 5,
            'random_state': RANDOM_SEED
        }

        # Visualization parameters
        self.viz_params = {
            'figure_size': (12, 6),
            'large_figure_size': (16, 10),
            'dpi': 100,
            'color_palette': 'viridis'
        }

        # Create all directories
        self._create_directories()

        # Initialize execution log
        self.execution_log = []
        self.log_file = f"{self.dirs['logs']}/execution_log_{datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"

    def _create_directories(self):
        """Create all required directories if they don't exist."""
        for dir_name, dir_path in self.dirs.items():
            os.makedirs(dir_path, exist_ok=True)
        print(f"[CONFIG] Directory structure created at: {self.base_path}")

    def log_execution(self, message: str):
        """
        Log execution message with timestamp.

        Args:
            message: Message to log
        """
        timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        log_entry = f"[{timestamp}] {message}"
        self.execution_log.append(log_entry)
        print(log_entry)

        # Write to log file
        with open(self.log_file, 'a') as f:
            f.write(log_entry + '\n')

    def get_cache_path(self, dataset_name: str, step: str) -> str:
        """
        Get cache file path for a specific dataset and processing step.

        Args:
            dataset_name: Name of the dataset
            step: Processing step name

        Returns:
            Full path to cache file
        """
        return f"{self.dirs['cache']}/{dataset_name}_{step}.pkl"

    def cache_exists(self, dataset_name: str, step: str) -> bool:
        """
        Check if cached artifact exists.

        Args:
            dataset_name: Name of the dataset
            step: Processing step name

        Returns:
            True if cache exists, False otherwise
        """
        cache_path = self.get_cache_path(dataset_name, step)
        return os.path.exists(cache_path)

    def save_to_cache(self, data: Any, dataset_name: str, step: str):
        """
        Save data to cache.

        Args:
            data: Data to cache
            dataset_name: Name of the dataset
            step: Processing step name
        """
        cache_path = self.get_cache_path(dataset_name, step)
        with open(cache_path, 'wb') as f:
            pickle.dump(data, f)
        self.log_execution(f"Cached {step} for {dataset_name} at {cache_path}")

    def load_from_cache(self, dataset_name: str, step: str) -> Any:
        """
        Load data from cache.

        Args:
            dataset_name: Name of the dataset
            step: Processing step name

        Returns:
            Cached data
        """
        cache_path = self.get_cache_path(dataset_name, step)
        with open(cache_path, 'rb') as f:
            data = pickle.load(f)
        self.log_execution(f"Loaded {step} for {dataset_name} from cache")
        return data


def mount_google_drive(config: PipelineConfig):
    """
    Configure Google Drive usage for an environment where Drive
    has already been mounted externally (e.g., via:
        from google.colab import drive
        drive.mount('/content/drive')
    ).

    This function will NOT attempt to mount Drive again. It will:
      - verify that the expected mount point exists,
      - keep using Google Drive paths if available,
      - otherwise gracefully fall back to local storage.

    Args:
        config: Pipeline configuration object
    """
    if not config.use_google_drive:
        print("[CONFIG] Google Drive integration disabled. Using local storage.")
        return

    if os.path.exists(config.drive_mount_point):
        print(f"[CONFIG] Detected existing Google Drive mount at {config.drive_mount_point}")
        print(f"[CONFIG] Using Google Drive base path: {config.base_path}")
    else:
        print("[CONFIG] Google Drive mount point not found. Falling back to local storage.")
        config.use_google_drive = False
        config.base_path = config.local_root
        config._create_directories()



# Initialize global configuration
# Set use_google_drive=True since Drive has already been mounted manually above
CONFIG = PipelineConfig(use_google_drive=True)
mount_google_drive(CONFIG)


print("\n" + "=" * 80)
print("CONFIGURATION SUMMARY")
print("=" * 80)
print(f"Base Path: {CONFIG.base_path}")
print(f"Google Drive: {'Enabled' if CONFIG.use_google_drive else 'Disabled'}")
print(f"Random Seed: {RANDOM_SEED}")
print(f"Test Size: {CONFIG.training_params['test_size']}")
print(f"Cross-Validation Folds: {CONFIG.training_params['cv_folds']}")
print("=" * 80 + "\n")
```

```python
import re
import nltk

# ============================================================================

class DataLoader:
    """
    Handles loading and initial validation of datasets with caching support.
    """

    def __init__(self, config: PipelineConfig):
        """
        Initialize DataLoader.

        Args:
            config: Pipeline configuration object
        """
        self.config = config
        self.loaded_datasets = {}

    def load_dataset(self, dataset_key: str, force_reload: bool = False) -> pd.DataFrame:
        """
        Load dataset with caching support.

        Args:
            dataset_key: Key identifying the dataset in config
            force_reload: If True, bypass cache and reload from source

        Returns:
            Loaded DataFrame
        """
        # Check cache first
        if not force_reload and self.config.cache_exists(dataset_key, 'raw'):
            self.config.log_execution(f"Loading {dataset_key} from cache")
            df = self.config.load_from_cache(dataset_key, 'raw')
            self.loaded_datasets[dataset_key] = df
            return df

        # Load from file
        dataset_info = self.config.datasets[dataset_key]
        filepath = dataset_info['filename']

        self.config.log_execution(f"Loading {dataset_key} from file: {filepath}")

        try:
            # Get separator from dataset_info, default to comma
            sep = dataset_info.get('sep', ',')
            df = pd.read_csv(filepath, sep=sep)

            # --- START DEBUGGING CODE ---
            # Removed this debugging print as the issue is identified and fixed.
            # if dataset_key == 'stress':
            #     print(f"[DEBUG] Columns found in {dataset_key} after loading: {df.columns.tolist()}")
            # --- END DEBUGGING CODE ---

            # Drop specified columns
            if dataset_info['drop_cols']:
                cols_to_drop = [col for col in dataset_info['drop_cols'] if col in df.columns]
                if cols_to_drop:
                    df = df.drop(columns=cols_to_drop)
                    self.config.log_execution(f"Dropped columns: {cols_to_drop}")

            # Basic validation
            assert not df.empty, f"Dataset {dataset_key} is empty"
            assert dataset_info['target'] in df.columns, f"Target column '{dataset_info['target']}' not found"

            # Cache the loaded data
            self.config.save_to_cache(df, dataset_key, 'raw')

            self.loaded_datasets[dataset_key] = df

            self.config.log_execution(
                f"Successfully loaded {dataset_key}: {df.shape[0]} rows, {df.shape[1]} columns"
            )

            return df

        except FileNotFoundError:
            self.config.log_execution(f"[ERROR] File not found: {filepath}")
            return None
        except Exception as e:
            self.config.log_execution(f"[ERROR] Failed to load {dataset_key}: {str(e)}")
            return None

    def load_all_datasets(self) -> Dict[str, pd.DataFrame]:
        """
        Load all configured datasets.

        Returns:
            Dictionary mapping dataset keys to DataFrames
        """
        self.config.log_execution("Starting to load all datasets...")

        for dataset_key in self.config.datasets.keys():
            self.load_dataset(dataset_key)

        self.config.log_execution(f"Loaded {len(self.loaded_datasets)} datasets successfully")

        return self.loaded_datasets

    def get_dataset_summary(self, dataset_key: str) -> Dict[str, Any]:
        """
        Get summary statistics for a dataset.

        Args:
            dataset_key: Key identifying the dataset

        Returns:
            Dictionary containing summary information
        """
        if dataset_key not in self.loaded_datasets:
            return {}

        df = self.loaded_datasets[dataset_key]
        dataset_info = self.config.datasets[dataset_key]

        summary = {
            'name': dataset_key,
            'shape': df.shape,
            'columns': list(df.columns),
            'target': dataset_info['target'],
            'type': dataset_info['type'],
            'task': dataset_info['task'],
            'missing_values': df.isnull().sum().sum(),
            'duplicate_rows': df.duplicated().sum(),
            'memory_usage_mb': df.memory_usage(deep=True).sum() / 1024 / 1024
        }

        # Target distribution
        if dataset_info['target'] in df.columns:
            summary['target_distribution'] = df[dataset_info['target']].value_counts().to_dict()

        return summary


class DataCleaner:
    """
    Handles data cleaning operations with type-specific logic.
    """

    def __init__(self, config: PipelineConfig):
        """
        Initialize DataCleaner.

        Args:
            config: Pipeline configuration object
        """
        self.config = config

    def clean_dataset(self, df: pd.DataFrame, dataset_key: str,
                     force_clean: bool = False) -> pd.DataFrame:
        """
        Clean dataset with caching support.

        Args:
            df: Input DataFrame
            dataset_key: Key identifying the dataset
            force_clean: If True, bypass cache and re-clean

        Returns:
            Cleaned DataFrame
        """
        # Check cache
        if not force_clean and self.config.cache_exists(dataset_key, 'cleaned'):
            return self.config.load_from_cache(dataset_key, 'cleaned')

        self.config.log_execution(f"Cleaning {dataset_key}...")

        df_clean = df.copy()
        dataset_info = self.config.datasets[dataset_key]

        # Remove duplicates
        initial_rows = len(df_clean)
        df_clean = df_clean.drop_duplicates()
        removed_dupes = initial_rows - len(df_clean)
        if removed_dupes > 0:
            self.config.log_execution(f"Removed {removed_dupes} duplicate rows")

        # Type-specific cleaning
        if dataset_info['type'] == 'text':
            df_clean = self._clean_text_dataset(df_clean, dataset_key)
        else:
            df_clean = self._clean_structured_dataset(df_clean, dataset_key)

        # Cache cleaned data
        self.config.save_to_cache(df_clean, dataset_key, 'cleaned')

        self.config.log_execution(
            f"Cleaning complete for {dataset_key}: {len(df_clean)} rows remaining"
        )

        return df_clean

    def _clean_text_dataset(self, df: pd.DataFrame, dataset_key: str) -> pd.DataFrame:
        """
        Clean text-based dataset (no encoding, only text processing prep).

        Args:
            df: Input DataFrame
            dataset_key: Dataset identifier

        Returns:
            Cleaned DataFrame
        """
        dataset_info = self.config.datasets[dataset_key]
        target_col = dataset_info['target']
        current_text_col_name = dataset_info.get('text_column') # This is 'combined_text_title' for bipolar

        # --- Specific handling for bipolar dataset: combine text and title ---
        if dataset_key == 'bipolar' and 'text' in df.columns and 'title' in df.columns:
            self.config.log_execution(f"Combining 'text' and 'title' into '{current_text_col_name}' for {dataset_key}")
            # Fill NaN values with empty string before combining
            df[current_text_col_name] = df['title'].fillna('') + ' ' + df['text'].fillna('')
            df = df.drop(columns=['text', 'title'])
            self.config.log_execution(f"Dropped original 'text' and 'title' columns for {dataset_key}")
            text_col_to_process = current_text_col_name # Use the newly created column
        elif current_text_col_name and current_text_col_name in df.columns:
            text_col_to_process = current_text_col_name
        else:
            # Fallback to previous inference logic if text_column is not defined or not found
            self.config.log_execution(f"[WARNING] 'text_column' not defined or not found for {dataset_key}. Attempting to infer.")
            inferred_text_cols = [col for col in df.columns if col != target_col and df[col].dtype == 'object']
            if inferred_text_cols:
                text_col_to_process = inferred_text_cols[0]
                self.config.log_execution(f"[INFO] Inferred text column for {dataset_key}: {text_col_to_process}")
            else:
                raise ValueError(f"No suitable text column found for {dataset_key}")

        # At this point, text_col_to_process holds the actual text column to use for cleaning.

        # Remove rows with missing text
        initial_rows = len(df)
        df = df.dropna(subset=[text_col_to_process])
        removed_missing = initial_rows - len(df)
        if removed_missing > 0:
            self.config.log_execution(f"Removed {removed_missing} rows with missing text")

        # Ensure the text column is string type before applying string operations
        df[text_col_to_process] = df[text_col_to_process].astype(str)

        # Remove empty strings
        df = df[df[text_col_to_process].str.strip() != '']

        # Remove rows with missing target
        df = df.dropna(subset=[target_col])

        return df

    def _clean_structured_dataset(self, df: pd.DataFrame, dataset_key: str) -> pd.DataFrame:
        """
        Clean structured dataset with numeric/categorical features.

        Args:
            df: Input DataFrame
            dataset_key: Dataset identifier

        Returns:
            Cleaned DataFrame
        """
        dataset_info = self.config.datasets[dataset_key]
        target_col = dataset_info['target']

        # Remove rows with missing target
        df = df.dropna(subset=[target_col])

        # Identify numeric and categorical columns
        numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
        categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

        # Remove target from feature lists
        if target_col in numeric_cols:
            numeric_cols.remove(target_col)
        if target_col in categorical_cols:
            categorical_cols.remove(target_col)

        # Handle missing values in numeric columns (median imputation)
        for col in numeric_cols:
            if df[col].isnull().sum() > 0:
                median_val = df[col].median()
                df[col].fillna(median_val, inplace=True)
                self.config.log_execution(f"Imputed {col} with median: {median_val:.2f}")

        # Handle missing values in categorical columns (mode imputation)
        for col in categorical_cols:
            if df[col].isnull().sum() > 0:
                mode_val = df[col].mode()[0] if not df[col].mode().empty else 'Unknown'
                df[col].fillna(mode_val, inplace=True)
                self.config.log_execution(f"Imputed {col} with mode: {mode_val}")

        return df


class TextPreprocessor:
    """
    Handles NLP preprocessing for text datasets.
    """

    def __init__(self, config: PipelineConfig):
        """
        Initialize TextPreprocessor.

        Args:
            config: Pipeline configuration object
        """
        self.config = config
        self.stop_words = set(stopwords.words('english'))
        self.lemmatizer = WordNetLemmatizer()
        self.stemmer = PorterStemmer()

    def preprocess_text(self, text: str, method: str = 'lemmatize') -> str:
        """
        Preprocess individual text string.

        Args:
            text: Input text
            method: 'lemmatize' or 'stem'

        Returns:
            Preprocessed text
        """
        # Convert to lowercase
        text = text.lower()

        # Remove URLs
        text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

        # Remove email addresses
        text = re.sub(r'\S+@\S+', '', text)

        # Remove mentions and hashtags
        text = re.sub(r'@\w+|#\w+', '', text)

        # Remove special characters and digits
        text = re.sub(r'[^a-zA-Z\s]', '', text)

        # Tokenize
        tokens = word_tokenize(text)

        # Remove stopwords and short tokens
        tokens = [token for token in tokens if token not in self.stop_words and len(token) > 2]

        # Lemmatize or stem
        if method == 'lemmatize':
            tokens = [self.lemmatizer.lemmatize(token) for token in tokens]
        elif method == 'stem':
            tokens = [self.stemmer.stem(token) for token in tokens]

        return ' '.join(tokens)

    def preprocess_text_column(self, df: pd.DataFrame, text_col: str,
                               dataset_key: str) -> pd.DataFrame:
        """
        Preprocess entire text column with caching.

        Args:
            df: Input DataFrame
            text_col: Name of text column
            dataset_key: Dataset identifier

        Returns:
            DataFrame with preprocessed text
        """
        # Check cache
        if self.config.cache_exists(dataset_key, 'preprocessed_text'):
            return self.config.load_from_cache(dataset_key, 'preprocessed_text')

        self.config.log_execution(f"Preprocessing text for {dataset_key}...")

        df_processed = df.copy()

        # Create processed text column
        df_processed['processed_text'] = df_processed[text_col].apply(
            lambda x: self.preprocess_text(str(x))
        )

        # Remove rows where preprocessing resulted in empty strings
        initial_rows = len(df_processed)
        df_processed = df_processed[df_processed['processed_text'].str.strip() != '']
        removed_empty = initial_rows - len(df_processed)
        if removed_empty > 0:
            self.config.log_execution(f"Removed {removed_empty} rows with empty processed text")

        # Cache
        self.config.save_to_cache(df_processed, dataset_key, 'preprocessed_text')

        self.config.log_execution(f"Text preprocessing complete for {dataset_key}")

        return df_processed

    def extract_text_features(self, df: pd.DataFrame, text_col: str) -> pd.DataFrame:
        """
        Extract basic text features (length, word count, etc.).

        Args:
            df: Input DataFrame
            text_col: Name of text column

        Returns:
            DataFrame with additional text feature columns
        """
        df_features = df.copy()

        # Character count
        df_features['char_count'] = df_features[text_col].apply(len)

        # Word count
        df_features['word_count'] = df_features[text_col].apply(lambda x: len(str(x).split()))

        # Average word length
        df_features['avg_word_length'] = df_features[text_col].apply(
            lambda x: np.mean([len(word) for word in str(x).split()]) if len(str(x).split()) > 0 else 0
        )

        # Sentence count
        df_features['sentence_count'] = df_features[text_col].apply(
            lambda x: len(sent_tokenize(str(x)))
        )

        # Uppercase ratio
        df_features['uppercase_ratio'] = df_features[text_col].apply(
            lambda x: sum(1 for c in str(x) if c.isupper()) / len(str(x)) if len(str(x)) > 0 else 0
        )

        return df_features


class VisualizationEngine:
    """
    Centralized visualization functions.
    """

    def __init__(self, config: PipelineConfig):
        """
        Initialize VisualizationEngine.

        Args:
            config: Pipeline configuration object
        """
        self.config = config

    def plot_target_distribution(self, df: pd.DataFrame, target_col: str,
                                 dataset_key: str):
        """
        Plot target variable distribution.

        Args:
            df: Input DataFrame
            target_col: Target column name
            dataset_key: Dataset identifier
        """
        fig, axes = plt.subplots(1, 2, figsize=self.config.viz_params['figure_size'])

        # Count plot
        target_counts = df[target_col].value_counts()
        axes[0].bar(range(len(target_counts)), target_counts.values, color='skyblue')
        axes[0].set_xticks(range(len(target_counts)))
        axes[0].set_xticklabels(target_counts.index, rotation=45)
        axes[0].set_xlabel('Class')
        axes[0].set_ylabel('Count')
        axes[0].set_title(f'Target Distribution - {dataset_key}')
        axes[0].grid(axis='y', alpha=0.3)

        # Pie chart
        axes[1].pie(target_counts.values, labels=target_counts.index, autopct='%1.1f%%',
                   colors=sns.color_palette('pastel'))
        axes[1].set_title(f'Target Proportion - {dataset_key}')

        plt.tight_layout()
        save_path = f"{self.config.dirs['plots_eda']}/{dataset_key}_target_distribution.png"
        plt.savefig(save_path, dpi=self.config.viz_params['dpi'], bbox_inches='tight')
        plt.close()

        self.config.log_execution(f"Saved target distribution plot: {save_path}")

    def plot_missing_data(self, df: pd.DataFrame, dataset_key: str):
        """
        Visualize missing data patterns.

        Args:
            df: Input DataFrame
            dataset_key: Dataset identifier
        """
        if not MISSINGNO_AVAILABLE:
            return

        fig, axes = plt.subplots(1, 2, figsize=self.config.viz_params['large_figure_size'])

        # Missing data matrix
        msno.matrix(df, ax=axes[0], sparkline=False)
        axes[0].set_title(f'Missing Data Matrix - {dataset_key}')

        # Missing data bar chart
        msno.bar(df, ax=axes[1])
        axes[1].set_title(f'Missing Data Count - {dataset_key}')

        plt.tight_layout()
        save_path = f"{self.config.dirs['plots_eda']}/{dataset_key}_missing_data.png"
        plt.savefig(save_path, dpi=self.config.viz_params['dpi'], bbox_inches='tight')
        plt.close()

        self.config.log_execution(f"Saved missing data plot: {save_path}")

    def plot_correlation_matrix(self, df: pd.DataFrame, dataset_key: str):
        """
        Plot correlation matrix for numeric features.

        Args:
            df: Input DataFrame
            dataset_key: Dataset identifier
        """
        numeric_cols = df.select_dtypes(include=[np.number]).columns
        if len(numeric_cols) < 2:
            return

        fig, ax = plt.subplots(figsize=self.config.viz_params['large_figure_size'])

        corr_matrix = df[numeric_cols].corr()
        sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm',
                   center=0, square=True, ax=ax, cbar_kws={"shrink": 0.8})
        ax.set_title(f'Correlation Matrix - {dataset_key}')

        plt.tight_layout()
        save_path = f"{self.config.dirs['plots_eda']}/{dataset_key}_correlation_matrix.png"
        plt.savefig(save_path, dpi=self.config.viz_params['dpi'], bbox_inches='tight')
        plt.close()

        self.config.log_execution(f"Saved correlation matrix: {save_path}")

    def plot_word_cloud(self, text_series: pd.Series, dataset_key: str, label: str = 'all'):
        """
        Generate and save word cloud.

        Args:
            text_series: Series containing text data
            dataset_key: Dataset identifier
            label: Label for this word cloud
        """
        if not WORDCLOUD_AVAILABLE:
            return

        # Combine all text
        text = ' '.join(text_series.astype(str).tolist())

        # Generate word cloud
        wordcloud = WordCloud(
            width=800, height=400,
            background_color='white',
            colormap='viridis',
            max_words=100
        ).generate(text)

        fig, ax = plt.subplots(figsize=self.config.viz_params['figure_size'])
        ax.imshow(wordcloud, interpolation='bilinear')
        ax.axis('off')
        ax.set_title(f'Word Cloud - {dataset_key} (label)')

        plt.tight_layout()
        save_path = f"{self.config.dirs['plots_eda']}/{dataset_key}_wordcloud_{label}.png"
        plt.savefig(save_path, dpi=self.config.viz_params['dpi'], bbox_inches='tight')
        plt.close()

        self.config.log_execution(f"Saved word cloud: {save_path}")

    def plot_confusion_matrix(self, y_true, y_pred, dataset_key: str, model_name: str, base_save_dir: Optional[str] = None):
        """
        Plot confusion matrix for classification results.

        Args:
            y_true: True labels
            y_pred: Predicted labels
            dataset_key: Dataset identifier
            model_name: Model name
            base_save_dir: Optional base directory to save the plot. If None, uses default plots_models directory.
        """
        cm = confusion_matrix(y_true, y_pred)

        fig, ax = plt.subplots(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax, annot_kws={"size": 25}) # Increased font size for annotations
        ax.set_xlabel('Predicted', fontsize=20) # Increased fontsize for axis labels
        ax.set_ylabel('Actual', fontsize=16)   # Increased fontsize for axis labels
        ax.set_title(f'Confusion Matrix - {model_name} on {dataset_key}')

        # Increase font size for tick labels (0 and 1)
        ax.tick_params(axis='x', labelsize=20) # Adjust labelsize as needed
        ax.tick_params(axis='y', labelsize=20) # Adjust labelsize as needed

        plt.tight_layout()

        # Determine the final save path
        if base_save_dir:
            os.makedirs(base_save_dir, exist_ok=True) # Ensure the custom directory exists
            final_save_path = f"{base_save_dir}/{dataset_key}_{model_name}_confusion_matrix.png"
        else:
            final_save_path = f"{self.config.dirs['plots_models']}/{dataset_key}_{model_name}_confusion_matrix.png"

        plt.savefig(final_save_path, dpi=self.config.viz_params['dpi'], bbox_inches='tight')
        plt.close()

        self.config.log_execution(f"Saved confusion matrix: {final_save_path}")

    def plot_roc_curve(self, y_true, y_pred_proba, dataset_key: str, model_name: str):
        """
        Plot ROC curve.

        Args:
            y_true: True labels
            y_pred_proba: Predicted probabilities
            dataset_key: Dataset identifier
            model_name: Model name
        """
        fpr, tpr, _ = roc_curve(y_true, y_pred_proba)
        auc_score = roc_auc_score(y_true, y_pred_proba)

        fig, ax = plt.subplots(figsize=(8, 6))
        ax.plot(fpr, tpr, label=f'ROC Curve (AUC = {auc_score:.3f})', linewidth=2)
        ax.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
        ax.set_xlabel('False Positive Rate')
        ax.set_ylabel('True Positive Rate')
        ax.set_title(f'ROC Curve - {model_name} on {dataset_key}')
        ax.legend()
        ax.grid(alpha=0.3)

        plt.tight_layout()
        save_path = f"{self.config.dirs['plots_models']}/{dataset_key}_{model_name}_roc_curve.png"
        plt.savefig(save_path, dpi=self.config.viz_params['dpi'], bbox_inches='tight')
        plt.close()

        self.config.log_execution(f"Saved ROC curve: {save_path}")
```

## FEATURE ENGINEERING AND SELECTION

```python
import os
import joblib
from scipy.sparse import hstack, csr_matrix
from sklearn.preprocessing import MinMaxScaler

# ============================================================================

class FeatureEngineer:

    def __init__(self, config: PipelineConfig):
        self.config = config
        self.scalers = {}
        self.encoders = {}
        self.tfidf_vectorizers = {}

    # ------------------------------------------------------------------------

    def engineer_structured_features(self, df, dataset_key, target_col):

        if self.config.cache_exists(dataset_key, "engineered_features"):
            print(f"{dataset_key}: loading engineered features from cache")
            cached = self.config.load_from_cache(dataset_key, "engineered_features")
            return cached["df"], cached["feature_cols"]

        print(f"Engineering structured features for {dataset_key}")

        df_eng = df.copy()

        numeric_cols = df_eng.select_dtypes(include=[np.number]).columns.tolist()
        categorical_cols = df_eng.select_dtypes(include=["object"]).columns.tolist()

        if target_col in numeric_cols:
            numeric_cols.remove(target_col)

        if target_col in categorical_cols:
            categorical_cols.remove(target_col)

        for col in categorical_cols:
            le = LabelEncoder()
            df_eng[f"{col}_encoded"] = le.fit_transform(df_eng[col].astype(str))
            self.encoders[f"{dataset_key}_{col}"] = le

        for col in numeric_cols[:5]:
            df_eng[f"{col}_squared"] = df_eng[col] ** 2

        feature_cols = [
            col for col in df_eng.columns
            if col != target_col and col not in categorical_cols
        ]

        scaler = StandardScaler()
        df_eng[feature_cols] = scaler.fit_transform(df_eng[feature_cols])

        cache_data = {
            "df": df_eng,
            "feature_cols": feature_cols
        }

        self.config.save_to_cache(cache_data, dataset_key, "engineered_features")

        return df_eng, feature_cols

    # ------------------------------------------------------------------------

    def create_tfidf_features(self, df, text_col, dataset_key):

        if self.config.cache_exists(dataset_key, "tfidf_features"):
            print(f"{dataset_key}: loading TF-IDF from cache")
            cached = self.config.load_from_cache(dataset_key, "tfidf_features")
            # Ensure vectorizer is stored in the object even if loaded from cache
            self.tfidf_vectorizers[dataset_key] = cached["vectorizer"]
            return cached["matrix"], cached["vectorizer"]

        print(f"Creating TF-IDF for {dataset_key}")

        vectorizer = TfidfVectorizer(
            max_features=self.config.nlp_params["max_features_tfidf"],
            ngram_range=self.config.nlp_params["ngram_range"],
            min_df=self.config.nlp_params["min_df"],
            max_df=self.config.nlp_params["max_df"]
        )

        tfidf_matrix = vectorizer.fit_transform(df[text_col])

        # store vectorizer for export
        self.tfidf_vectorizers[dataset_key] = vectorizer

        cache_data = {
            "matrix": tfidf_matrix,
            "vectorizer": vectorizer
        }

        self.config.save_to_cache(cache_data, dataset_key, "tfidf_features")

        return tfidf_matrix, vectorizer

# ============================================================================

feature_engineer = FeatureEngineer(CONFIG)

engineered_datasets = {}
feature_columns = {}

# ---------------------------------------------------------------------------
# STRUCTURED DATASETS
# ---------------------------------------------------------------------------

print("\n>>> Structured Feature Engineering <<<")

for dataset_key in []:

    if dataset_key not in cleaned_datasets:
        continue

    df = cleaned_datasets[dataset_key]
    target = CONFIG.datasets[dataset_key]["target"]

    df_eng, feat_cols = feature_engineer.engineer_structured_features(
        df, dataset_key, target
    )

    engineered_datasets[dataset_key] = df_eng
    feature_columns[dataset_key] = feat_cols

    print(dataset_key, "features:", len(feat_cols))

# ---------------------------------------------------------------------------
# TEXT DATASETS
# ---------------------------------------------------------------------------

print("\n>>> Text Feature Engineering <<<")

from scipy.sparse import hstack, csr_matrix
from sklearn.preprocessing import MinMaxScaler

text_feature_matrices = {}

for dataset_key in ["anxiety", "depression", "suicide","stress","bipolar"]:

    if dataset_key not in preprocessed_text_datasets:
        continue

    df = preprocessed_text_datasets[dataset_key]

    tfidf_matrix, vectorizer = feature_engineer.create_tfidf_features(
        df, "processed_text", dataset_key
    )

    text_features = [
        "char_count",
        "word_count",
        "avg_word_length",
        "sentence_count",
        "uppercase_ratio"
    ]

    numeric = df[text_features].astype("float32").values

    scaler = MinMaxScaler()
    numeric = scaler.fit_transform(numeric)

    # Save scaler for export later
    feature_engineer.scalers[f"{dataset_key}_text_scaler"] = scaler

    numeric_sparse = csr_matrix(numeric)

    X = hstack([tfidf_matrix, numeric_sparse])

    y = df[CONFIG.datasets[dataset_key]["target"]].values

    if y.dtype == "object":
        y = LabelEncoder().fit_transform(y)

    text_feature_matrices[dataset_key] = {"X": X, "y": y}

    print(dataset_key, "matrix shape:", X.shape)

# ---------------------------------------------------------------------------
# FEATURE SELECTION (CACHE ENABLED)
# ---------------------------------------------------------------------------

print("\n>>> Feature Selection <<<")

selected_features = {}

for dataset_key, data in text_feature_matrices.items():

    if CONFIG.cache_exists(dataset_key, "chi2_features"):
        print(dataset_key, ": loading selected features from cache")
        selected_features[dataset_key] = CONFIG.load_from_cache(dataset_key, "chi2_features")
        continue

    X = data["X"]
    y = data["y"]

    if X.shape[0] > 200000:
        k = 500
    else:
        k = 1000

    selector = SelectKBest(chi2, k=min(k, X.shape[1]))

    X_selected = selector.fit_transform(X, y)

    selected_features[dataset_key] = {
        "X": X_selected,
        "y": y,
        "selector": selector
    }

    CONFIG.save_to_cache(selected_features[dataset_key], dataset_key, "chi2_features")

    print(dataset_key, "selected shape:", X_selected.shape)

# ---------------------------------------------------------------------------
# DIMENSIONALITY REDUCTION (CACHE ENABLED)
# ---------------------------------------------------------------------------

print("\n>>> Dimensionality Reduction <<<")

from sklearn.decomposition import TruncatedSVD

for dataset_key, data in selected_features.items():

    if CONFIG.cache_exists(dataset_key, "svd_features"):
        print(dataset_key, ": loading SVD from cache")
        svd_data = CONFIG.load_from_cache(dataset_key, "svd_features")
        continue

    X = data["X"]

    svd = TruncatedSVD(n_components=50, random_state=RANDOM_SEED)

    X_reduced = svd.fit_transform(X)

    svd_data = {
        "X_reduced": X_reduced,
        "svd_model": svd
    }

    CONFIG.save_to_cache(svd_data, dataset_key, "svd_features")

    print(dataset_key, "reduced shape:", X_reduced.shape)

print("\n" + "=" * 80)
print("STEP 4 COMPLETE")
print("=" * 80 + "\n")
```

## CLUSTERING AND ANOMALY DETECTION

```python
# ============================================================================

print("\n" + "=" * 80)
print("STEP 5: CLUSTERING AND ANOMALY DETECTION")
print("=" * 80 + "\n")

from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.decomposition import TruncatedSVD
from collections import Counter
import numpy as np


class ClusteringAnalyzer:

    def __init__(self, config):
        self.config = config

    # ------------------------------------------------------------------
    # KMEANS CLUSTERING
    # ------------------------------------------------------------------

    def perform_kmeans_clustering(self, X, dataset_key, n_clusters=5):

        cache_key = "kmeans_clusters"

        if self.config.cache_exists(dataset_key, cache_key):
            print(f"{dataset_key}: loading KMeans clusters from cache")
            return self.config.load_from_cache(dataset_key, cache_key)

        self.config.log_execution(f"Running KMeans for {dataset_key}")

        kmeans = KMeans(n_clusters=n_clusters, random_state=RANDOM_SEED)

        cluster_labels = kmeans.fit_predict(X)

        counts = Counter(cluster_labels)

        print(f"\n{dataset_key} KMeans (k={n_clusters})")

        for cid, count in sorted(counts.items()):
            print(f"Cluster {cid}: {count}")

        self.config.save_to_cache(cluster_labels, dataset_key, cache_key)

        return cluster_labels

    # ------------------------------------------------------------------
    # HIERARCHICAL CLUSTERING
    # ------------------------------------------------------------------

    def perform_hierarchical_clustering(self, X, dataset_key, n_clusters=4):

        cache_key = "hierarchical_clusters"

        if self.config.cache_exists(dataset_key, cache_key):
            print(f"{dataset_key}: loading hierarchical clusters from cache")
            return self.config.load_from_cache(dataset_key, cache_key)

        self.config.log_execution(f"Running hierarchical clustering for {dataset_key}")

        if X.shape[0] > 1000:
            idx = np.random.choice(X.shape[0], 1000, replace=False)
            X_sample = X[idx]
        else:
            X_sample = X

        if hasattr(X_sample, "toarray"):
            X_sample = X_sample.toarray()

        model = AgglomerativeClustering(n_clusters=n_clusters)

        labels = model.fit_predict(X_sample)

        self.config.save_to_cache(labels, dataset_key, cache_key)

        return labels

    # ------------------------------------------------------------------
    # ISOLATION FOREST
    # ------------------------------------------------------------------

    def detect_anomalies_isolation_forest(self, X, dataset_key, contamination=0.1):

        cache_key = "isolation_forest"

        if self.config.cache_exists(dataset_key, cache_key):
            print(f"{dataset_key}: loading IsolationForest from cache")
            return self.config.load_from_cache(dataset_key, cache_key)

        self.config.log_execution(f"Running IsolationForest for {dataset_key}")

        model = IsolationForest(
            contamination=contamination,
            random_state=RANDOM_SEED
        )

        labels = model.fit_predict(X)

        anomalies = (labels == -1).sum()

        print(f"{dataset_key} anomalies detected: {anomalies}")

        self.config.save_to_cache(labels, dataset_key, cache_key)

        return labels

    # ------------------------------------------------------------------
    # LOCAL OUTLIER FACTOR
    # ------------------------------------------------------------------

    def detect_anomalies_lof(self, X, dataset_key, contamination=0.1):

        cache_key = "lof_anomalies"

        if self.config.cache_exists(dataset_key, cache_key):
            print(f"{dataset_key}: loading LOF results from cache")
            return self.config.load_from_cache(dataset_key, cache_key)

        self.config.log_execution(f"Running LOF for {dataset_key}")

        if X.shape[0] > 5000:
            idx = np.random.choice(X.shape[0], 5000, replace=False)
            X_sample = X[idx]
        else:
            X_sample = X

        model = LocalOutlierFactor(contamination=contamination)

        labels = model.fit_predict(X_sample)

        anomalies = (labels == -1).sum()

        print(f"{dataset_key} LOF anomalies: {anomalies}")

        self.config.save_to_cache(labels, dataset_key, cache_key)

        return labels


# ============================================================================
# Helper function for visualization (replacement for apply_pca)
# ============================================================================

def reduce_dimensions(X, n_components=2):

    svd = TruncatedSVD(n_components=n_components, random_state=RANDOM_SEED)

    X_reduced = svd.fit_transform(X)

    return X_reduced


# ============================================================================
# RUN CLUSTERING
# ============================================================================

clustering_analyzer = ClusteringAnalyzer(CONFIG)

print("\n>>> KMeans Clustering <<<")

for dataset_key in ["anxiety", "depression", "suicide","stress","bipolar"]:

    if dataset_key not in text_feature_matrices:
        continue

    X = text_feature_matrices[dataset_key]["X"]

    clusters = clustering_analyzer.perform_kmeans_clustering(
        X,
        dataset_key,
        n_clusters=5
    )

    # dimensionality reduction for visualization
    X_reduced = reduce_dimensions(X, 2)

    fig, ax = plt.subplots(figsize=(10,6))

    scatter = ax.scatter(
        X_reduced[:,0],
        X_reduced[:,1],
        c=clusters,
        cmap="tab10",
        alpha=0.6
    )

    plt.colorbar(scatter)

    ax.set_xlabel("Component 1")
    ax.set_ylabel("Component 2")
    ax.set_title(f"KMeans Clusters - {dataset_key}")

    save_path = f"{CONFIG.dirs['plots_eda']}/{dataset_key}_kmeans_clusters.png"

    plt.savefig(save_path, dpi=CONFIG.viz_params['dpi'], bbox_inches="tight")

    plt.close()

    CONFIG.log_execution(f"Saved KMeans plot: {save_path}")


# ============================================================================
# HIERARCHICAL CLUSTERING
# ============================================================================

print("\n>>> Hierarchical Clustering <<<")

for dataset_key in ["anxiety", "depression"]:

    if dataset_key not in text_feature_matrices:
        continue

    X = text_feature_matrices[dataset_key]["X"]

    clustering_analyzer.perform_hierarchical_clustering(X, dataset_key)


# ============================================================================
# ANOMALY DETECTION
# ============================================================================

print("\n>>> Anomaly Detection <<<")

for dataset_key in ["anxiety", "depression", "suicide","stress","bipolar"]:

    if dataset_key not in text_feature_matrices:
        continue

    X = text_feature_matrices[dataset_key]["X"]

    clustering_analyzer.detect_anomalies_isolation_forest(X, dataset_key)

    clustering_analyzer.detect_anomalies_lof(X, dataset_key)


print("\n" + "=" * 80)
print("STEP 5 COMPLETE: Clustering and anomaly detection finished")
print("=" * 80 + "\n")
```

## HANDLING CLASS IMBALANCE

```python
# ============================================================================
print("\n" + "=" * 80)
print("STEP 6: HANDLING CLASS IMBALANCE")
print("=" * 80 + "\n")

import os
import joblib
from collections import Counter

from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from scipy.sparse import hstack, csr_matrix

# ============================================================================
# Imbalance Handler
# ============================================================================

class ImbalanceHandler:

    def __init__(self, config):
        self.config = config

    def apply_smote(self, X, y, dataset_key):

        if not IMBALANCED_AVAILABLE:
            print("[WARNING] SMOTE not available")
            return X, y

        print(f"\nApplying SMOTE for {dataset_key}")

        class_counts = Counter(y)

        print("Original distribution:")
        for k, v in class_counts.items():
            print(f"  Class {k}: {v}")

        # ------------------------------------------------------------------
        # Skip SMOTE for huge datasets (prevents RAM crash)
        # ------------------------------------------------------------------

        if X.shape[0] > 150000:
            print(f"[INFO] Skipping SMOTE for {dataset_key} (dataset too large)")
            return X, y

        # ------------------------------------------------------------------
        # Check minority class size
        # ------------------------------------------------------------------

        if len(class_counts) < 2 or min(class_counts.values()) < 2:
            print(f"[WARNING] SMOTE not applicable for {dataset_key}")
            return X, y

        try:

            smote = SMOTE(random_state=RANDOM_SEED)

            # IMPORTANT: keep sparse matrix (do NOT convert to dense)
            X_resampled, y_resampled = smote.fit_resample(X, y)

            new_counts = Counter(y_resampled)

            print("After SMOTE:")
            for k, v in new_counts.items():
                print(f"  Class {k}: {v}")

            return X_resampled, y_resampled

        except Exception as e:

            print(f"[WARNING] SMOTE failed for {dataset_key}: {e}")
            return X, y


# ============================================================================
# Initialize
# ============================================================================

imbalance_handler = ImbalanceHandler(CONFIG)

all_datasets_features = {}

CACHE_DIR = CONFIG.dirs["cache"]
os.makedirs(CACHE_DIR, exist_ok=True)


# ============================================================================
# Prepare TEXT DATASETS
# ============================================================================

print("\n>>> Preparing Text Dataset Features <<<")

for dataset_key in ["anxiety", "depression", "suicide","stress","bipolar"]:

    if dataset_key not in text_feature_matrices:
        continue

    X = text_feature_matrices[dataset_key]["X"]
    y = text_feature_matrices[dataset_key]["y"]

    all_datasets_features[dataset_key] = {"X": X, "y": y}

    print(f"{dataset_key}: X={X.shape}, y={y.shape}")


# ============================================================================
# Prepare STRUCTURED DATASETS
# ============================================================================

print("\n>>> Preparing Structured Dataset Features <<<")

for dataset_key in []:

    if dataset_key not in cleaned_datasets:
        continue

    df = cleaned_datasets[dataset_key].copy()

    target_col = CONFIG.datasets[dataset_key]["target"]

    y = df[target_col].values
    X_df = df.drop(columns=[target_col])

    numerical_cols = X_df.select_dtypes(include="number").columns
    categorical_cols = X_df.select_dtypes(include="object").columns


    # ----------------------------
    # One-hot encoding
    # ----------------------------

    if len(categorical_cols) > 0:

        encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=True)

        X_cat = encoder.fit_transform(X_df[categorical_cols])

        CONFIG.save_to_cache(encoder, dataset_key, "onehot_encoder")

    else:
        X_cat = csr_matrix((X_df.shape[0], 0))


    # ----------------------------
    # Scale numeric features
    # ----------------------------

    if len(numerical_cols) > 0:

        scaler = StandardScaler()

        X_num = scaler.fit_transform(X_df[numerical_cols])

        X_num = csr_matrix(X_num)

        CONFIG.save_to_cache(scaler, dataset_key, "standard_scaler")

    else:
        X_num = csr_matrix((X_df.shape[0], 0))


    # Combine features
    X = hstack([X_num, X_cat])


    # ----------------------------
    # Encode target
    # ----------------------------

    if y.dtype == "object" or len(set(y)) > 2:

        le = LabelEncoder()

        y = le.fit_transform(y)

        CONFIG.save_to_cache(le, dataset_key, "label_encoder")


    all_datasets_features[dataset_key] = {"X": X, "y": y}

    print(f"{dataset_key}: X={X.shape}, y={y.shape}")


# ============================================================================
# Apply SMOTE with CACHE
# ============================================================================

print("\n>>> Applying SMOTE to Datasets <<<")

balanced_datasets = {}

for dataset_key, data in all_datasets_features.items():

    cache_file = f"{CACHE_DIR}/{dataset_key}_balanced.pkl"


    # ------------------------------------------------
    # LOAD CACHE
    # ------------------------------------------------

    if os.path.exists(cache_file):

        print(f"\nLoading balanced dataset from cache: {dataset_key}")

        cached = joblib.load(cache_file)

        balanced_datasets[dataset_key] = cached

        print(
            f"{dataset_key} balanced: "
            f"X={cached['X'].shape}, "
            f"y={cached['y'].shape}"
        )

        continue


    # ------------------------------------------------
    # APPLY SMOTE
    # ------------------------------------------------

    X = data["X"]
    y = data["y"]

    X_balanced, y_balanced = imbalance_handler.apply_smote(X, y, dataset_key)


    balanced_datasets[dataset_key] = {
        "X": X_balanced,
        "y": y_balanced
    }


    # ------------------------------------------------
    # SAVE CACHE
    # ------------------------------------------------

    joblib.dump(balanced_datasets[dataset_key], cache_file)

    print(
        f"{dataset_key} balanced: "
        f"X={X_balanced.shape}, "
        f"y={y_balanced.shape}"
    )


print("\n" + "=" * 80)
print("STEP 6 COMPLETE: Class imbalance handling finished")
print("=" * 80 + "\n")
```

## MODEL TRAINING AND EVALUATION

```python
# ============================================================================

print("\n" + "=" * 80)
print("STEP 7: MODEL TRAINING AND EVALUATION")
print("=" * 80 + "\n")

import os
import joblib

class ModelTrainer:
    """
    Handles model training, evaluation, and persistence.
    """

    def __init__(self, config: PipelineConfig):
        self.config = config
        self.trained_models = {}
        self.evaluation_results = {}
        self.test_splits = {} # Added to store X_test and y_test for explainability

    # ------------------------------------------------------------------------
    # TRAIN TEST SPLIT
    # ------------------------------------------------------------------------
    def train_test_split_data(self, X, y):

        return train_test_split(
            X,
            y,
            test_size=self.config.training_params['test_size'],
            random_state=self.config.training_params['random_state'],
            stratify=y if len(np.unique(y)) > 1 else None
        )

    # ------------------------------------------------------------------------
    # CHECK / LOAD EXISTING MODEL
    # ------------------------------------------------------------------------
    def load_existing_model(self, dataset_key, model_name, expected_features=None):

        model_path = f"{self.config.dirs['models']}/{dataset_key}_{model_name}.pkl"

        if os.path.exists(model_path):
            print(f"Loading existing model: {model_name}")
            model = joblib.load(model_path)
            # Check for feature compatibility
            if expected_features is not None and hasattr(model, 'n_features_in_'):
                if model.n_features_in_ != expected_features:
                    self.config.log_execution(f"[WARNING] Loaded model {model_name} for {dataset_key} has {model.n_features_in_} features, but expected {expected_features}. Forcing retraining.")
                    return None # Incompatible model, force retraining
            return model

        return None

    # ------------------------------------------------------------------------
    # SAVE MODEL
    # ------------------------------------------------------------------------
    def save_model(self, model, dataset_key, model_name):

        model_path = f"{self.config.dirs['models']}/{dataset_key}_{model_name}.pkl"

        joblib.dump(model, model_path)

        self.config.log_execution(f"Saved model: {model_path}")

    # ------------------------------------------------------------------------
    # MODEL TRAINING FUNCTIONS
    # ------------------------------------------------------------------------
    def train_logistic_regression(self, X_train, y_train):

        model = LogisticRegression(
            max_iter=1000,
            random_state=RANDOM_SEED,
            n_jobs=-1
        )

        model.fit(X_train, y_train)

        return model

    def train_random_forest(self, X_train, y_train):

        model = RandomForestClassifier(
            n_estimators=200,
            max_depth=20,
            random_state=RANDOM_SEED,
            n_jobs=-1
        )

        model.fit(X_train, y_train)

        return model

    def train_gradient_boosting(self, X_train, y_train):

        model = GradientBoostingClassifier(
            n_estimators=100,
            learning_rate=0.1,
            max_depth=5,
            random_state=RANDOM_SEED
        )

        model.fit(X_train, y_train)

        return model

    def train_xgboost(self, X_train, y_train):

        if not XGBOOST_AVAILABLE:
            return None

        model = xgb.XGBClassifier(
            n_estimators=100,
            max_depth=5,
            learning_rate=0.1,
            random_state=RANDOM_SEED,
            n_jobs=-1,
            use_label_encoder=False,
            eval_metric="logloss"
        )

        model.fit(X_train, y_train)

        return model

    def train_lightgbm(self, X_train, y_train):

        if not LIGHTGBM_AVAILABLE:
            return None

        model = lgb.LGBMClassifier(
            n_estimators=100,
            max_depth=5,
            learning_rate=0.1,
            random_state=RANDOM_SEED,
            n_jobs=-1,
            verbose=-1
        )

        model.fit(X_train, y_train)

        return model

    # ------------------------------------------------------------------------
    # EVALUATION
    # ------------------------------------------------------------------------
    def evaluate_model(self, model, X_test, y_test, dataset_key, model_name):

        y_pred = model.predict(X_test)

        accuracy = accuracy_score(y_test, y_pred)

        precision = precision_score(
            y_test,
            y_pred,
            average="weighted",
            zero_division=0
        )

        recall = recall_score(
            y_test,
            y_pred,
            average="weighted",
            zero_division=0
        )

        f1 = f1_score(
            y_test,
            y_pred,
            average="weighted",
            zero_division=0
        )

        metrics = {
            "dataset": dataset_key,
            "model": model_name,
            "accuracy": accuracy,
            "precision": precision,
            "recall": recall,
            "f1_score": f1
        }

        if len(np.unique(y_test)) == 2:
            try:
                y_prob = model.predict_proba(X_test)[:, 1]
                roc = roc_auc_score(y_test, y_prob)
                metrics["roc_auc"] = roc
            except:
                pass

        print(f"\n{dataset_key} - {model_name}")
        print(f"Accuracy: {accuracy:.4f}")
        print(f"Precision: {precision:.4f}")
        print(f"Recall: {recall:.4f}")
        print(f"F1: {f1:.4f}")

        return metrics

    # ------------------------------------------------------------------------
    # TRAIN + SKIP EXISTING MODELS
    # ------------------------------------------------------------------------
    def train_and_evaluate_all(self, X, y, dataset_key):

        X_train, X_test, y_train, y_test = self.train_test_split_data(X, y)
        self.test_splits[dataset_key] = {'X_test': X_test, 'y_test': y_test} # Store test splits

        models = {}
        metrics_list = []

        model_registry = {
            "logistic_regression": self.train_logistic_regression,
            "random_forest": self.train_random_forest,
            "gradient_boosting": self.train_gradient_boosting,
            "xgboost": self.train_xgboost,
            "lightgbm": self.train_lightgbm
        }

        for model_name, train_func in model_registry.items():

            if model_name == "xgboost" and not XGBOOST_AVAILABLE:
                continue

            if model_name == "lightgbm" and not LIGHTGBM_AVAILABLE:
                continue

            # Pass expected_features to load_existing_model for compatibility check
            model = self.load_existing_model(dataset_key, model_name, expected_features=X_train.shape[1])

            if model is None:

                print(f"Training model: {model_name}")

                model = train_func(X_train, y_train)

                if model is None:
                    continue

                self.save_model(model, dataset_key, model_name)

            else:
                print(f"Skipping training (model exists): {model_name}")

            models[model_name] = model

            metrics = self.evaluate_model(
                model,
                X_test,
                y_test,
                dataset_key,
                model_name
            )

            metrics_list.append(metrics)

        self.trained_models[dataset_key] = models
        return {
            "models": models,
            "metrics": metrics_list,
            "X_test": X_test,
            "y_test": y_test
        }


# ----------------------------------------------------------------------------
# INITIALIZE TRAINER
# ----------------------------------------------------------------------------

model_trainer = ModelTrainer(CONFIG)

all_metrics = []

# ----------------------------------------------------------------------------
# TEXT DATASETS
# ----------------------------------------------------------------------------

print("\n>>> Training Models for Text Datasets <<<")

for dataset_key in ["anxiety", "depression", "suicide","stress","bipolar"]:

    if dataset_key not in balanced_datasets:
        continue

    print(f"\n{'='*60}")
    print(f"TRAINING MODELS FOR: {dataset_key.upper()}")
    print("="*60)

    X = balanced_datasets[dataset_key]["X"]
    y = balanced_datasets[dataset_key]["y"]

    result = model_trainer.train_and_evaluate_all(X, y, dataset_key)

    all_metrics.extend(result["metrics"])


# ----------------------------------------------------------------------------
# STRUCTURED DATASETS
# ----------------------------------------------------------------------------

print("\n>>> Training Models for Structured Datasets <<<")

for dataset_key in []:

    if dataset_key not in balanced_datasets:
        continue

    print(f"\n{'='*60}")
    print(f"TRAINING MODELS FOR: {dataset_key.upper()}")
    print("="*60)

    X = balanced_datasets[dataset_key]["X"]
    y = balanced_datasets[dataset_key]["y"]

    result = model_trainer.train_and_evaluate_all(X, y, dataset_key)

    all_metrics.extend(result["metrics"])


# ----------------------------------------------------------------------------
# SAVE METRICS
# ----------------------------------------------------------------------------

metrics_df = pd.DataFrame(all_metrics)

metrics_path = f"{CONFIG.dirs['reports']}/model_performance.csv"

metrics_df.to_csv(metrics_path, index=False)

CONFIG.log_execution(f"Saved model metrics: {metrics_path}")

print("\nMODEL PERFORMANCE SUMMARY\n")

print(metrics_df.to_string(index=False))
```

## MODEL EXPLAINABILITY

```python
# ============================================================================

print("\n" + "=" * 80)
print("STEP 8: MODEL EXPLAINABILITY")
print("=" * 80 + "\n")

import os
import numpy as np
import matplotlib.pyplot as plt

class ExplainabilityEngine:
    """
    Handles model explainability using SHAP and LIME.
    """

    def __init__(self, config):
        self.config = config

    # -----------------------------------------------------------------------
    # SHAP EXPLANATIONS
    # -----------------------------------------------------------------------

    def explain_with_shap(self, model, X_test, dataset_key, model_name):

        if not SHAP_AVAILABLE:
            self.config.log_execution("[WARNING] SHAP not available, skipping")
            return

        save_path = f"{self.config.dirs['plots_explainability']}/{dataset_key}_{model_name}_shap_summary.png"

        # Skip if explanation already exists
        if os.path.exists(save_path):
            self.config.log_execution(
                f"[SKIP] SHAP explanation already exists for {dataset_key} - {model_name}"
            )
            return

        self.config.log_execution(
            f"Generating SHAP explanations for {dataset_key} - {model_name}..."
        )

        try:

            # Sample for performance
            if X_test.shape[0] > 500:
                idx = np.random.choice(X_test.shape[0], 500, replace=False)
                X_sample = X_test[idx]
            else:
                X_sample = X_test

            # Convert sparse → dense for SHAP plotting
            if hasattr(X_sample, "toarray"):
                X_sample = X_sample.toarray()

            explainer = shap.TreeExplainer(model)

            shap_values_raw = explainer.shap_values(X_sample)

            if isinstance(shap_values_raw, list):
                shap_values = np.mean(
                    np.array([np.abs(v) for v in shap_values_raw]), axis=0
                )
            else:
                shap_values = shap_values_raw

            plt.figure(figsize=(12, 8))

            shap.summary_plot(
                shap_values,
                X_sample,
                show=False,
                max_display=20
            )

            plt.tight_layout()

            plt.savefig(
                save_path,
                dpi=self.config.viz_params["dpi"],
                bbox_inches="tight"
            )

            plt.close()

            self.config.log_execution(f"Saved SHAP summary plot: {save_path}")

        except Exception as e:

            self.config.log_execution(
                f"[WARNING] SHAP explanation failed for {dataset_key} - {model_name}: {str(e)}"
            )

    # -----------------------------------------------------------------------
    # LIME EXPLANATIONS FOR TEXT
    # -----------------------------------------------------------------------

    def explain_with_lime_text(
        self,
        model,
        text_data,
        dataset_key,
        model_name,
        vectorizer,
        idx=0
    ):

        if not LIME_AVAILABLE:
            return

        save_path = f"{self.config.dirs['plots_explainability']}/{dataset_key}_{model_name}_lime_text.png"

        # Skip if already exists
        if os.path.exists(save_path):
            self.config.log_execution(
                f"[SKIP] LIME explanation already exists for {dataset_key} - {model_name}"
            )
            return

        self.config.log_execution(
            f"Generating LIME explanation for {dataset_key} - {model_name}..."
        )

        try:

            class_names = ["Negative", "Positive"]

            y_test_unique = np.unique(
                model_trainer.test_splits[dataset_key]["y_test"]
            )

            if len(y_test_unique) == 2:
                class_names = [
                    f"Class {y_test_unique[0]}",
                    f"Class {y_test_unique[1]}"
                ]
            elif len(y_test_unique) > 2:
                class_names = [
                    f"Class {i}" for i in sorted(y_test_unique)
                ]

            explainer = LimeTextExplainer(class_names=class_names)

            def predict_proba_fn(texts):
                text_vectors = vectorizer.transform(texts)
                return model.predict_proba(text_vectors)

            sample_text = text_data.iloc[idx]

            exp = explainer.explain_instance(
                sample_text,
                predict_proba_fn,
                num_features=10
            )

            fig = exp.as_pyplot_figure()

            plt.tight_layout()

            plt.savefig(
                save_path,
                dpi=self.config.viz_params["dpi"],
                bbox_inches="tight"
            )

            plt.close()

            self.config.log_execution(
                f"Saved LIME text explanation: {save_path}"
            )

        except Exception as e:

            self.config.log_execution(
                f"[WARNING] LIME explanation failed for {dataset_key} - {model_name}: {str(e)}"
            )


# ============================================================================

explainability_engine = ExplainabilityEngine(CONFIG)

# ---------------------------------------------------------------------------
# SHAP EXPLANATIONS
# ---------------------------------------------------------------------------

print("\n>>> Generating SHAP Explanations <<<")

for dataset_key in model_trainer.trained_models.keys():

    models = model_trainer.trained_models[dataset_key]

    dataset_metrics = [
        m for m in all_metrics
        if m["dataset"] == dataset_key
    ]

    if not dataset_metrics:
        continue

    best_metric = max(
        dataset_metrics,
        key=lambda x: x["f1_score"]
    )

    best_model_name = best_metric["model"]

    if best_model_name not in models:
        continue

    best_model = models[best_model_name]

    X_test = model_trainer.test_splits[dataset_key]["X_test"]

    if best_model_name in [
        "random_forest",
        "gradient_boosting",
        "xgboost",
        "lightgbm"
    ]:

        explainability_engine.explain_with_shap(
            best_model,
            X_test,
            dataset_key,
            best_model_name
        )

# ---------------------------------------------------------------------------
# LIME EXPLANATIONS (TEXT DATASETS)
# ---------------------------------------------------------------------------

print("\n>>> Generating LIME Explanations (Text) <<<")

for dataset_key in ["anxiety", "depression", "suicide","stress","bipolar"]:

    if dataset_key not in preprocessed_text_datasets:
        continue

    if dataset_key not in model_trainer.trained_models:
        continue

    models = model_trainer.trained_models[dataset_key]

    dataset_metrics = [
        m for m in all_metrics
        if m["dataset"] == dataset_key
    ]

    if not dataset_metrics:
        continue

    best_metric = max(
        dataset_metrics,
        key=lambda x: x["f1_score"]
    )

    best_model_name = best_metric["model"]

    if best_model_name not in models:
        continue

    best_model = models[best_model_name]

    df_text = preprocessed_text_datasets[dataset_key]

    if dataset_key in feature_engineer.tfidf_vectorizers:

        vectorizer = feature_engineer.tfidf_vectorizers[dataset_key]

        explainability_engine.explain_with_lime_text(
            best_model,
            df_text["processed_text"],
            dataset_key,
            best_model_name,
            vectorizer,
            idx=0
        )

print("\n" + "=" * 80)
print("STEP 8 COMPLETE: Model explainability analysis finished")
print("=" * 80 + "\n")
```

## CROSS-VALIDATION AND HYPERPARAMETER TUNING

```python
# ============================================================================

print("\n" + "=" * 80)
print("STEP 9: CROSS-VALIDATION AND HYPERPARAMETER TUNING")
print("=" * 80 + "\n")

import os
import joblib
import pandas as pd
import numpy as np

from sklearn.model_selection import StratifiedKFold, cross_val_score, GridSearchCV
from sklearn.ensemble import RandomForestClassifier


class HyperparameterTuner:

    def __init__(self, config):
        self.config = config
        self.best_params = {}

    # ------------------------------------------------------------------
    # CROSS VALIDATION
    # ------------------------------------------------------------------

    def perform_cross_validation(self, model, X, y, dataset_key, model_name):

        self.config.log_execution(
            f"Running CV for {dataset_key} - {model_name}"
        )

        cv = StratifiedKFold(
            n_splits=self.config.training_params["cv_folds"],
            shuffle=True,
            random_state=RANDOM_SEED
        )

        scores = cross_val_score(
            model,
            X,
            y,
            cv=cv,
            scoring="f1_weighted",
            n_jobs=-1
        )

        print(f"\n{dataset_key} - {model_name}")
        print(f"Mean F1: {scores.mean():.4f}")
        print("Folds:", [f"{s:.4f}" for s in scores])

        return scores


    # ------------------------------------------------------------------
    # RANDOM FOREST TUNING
    # ------------------------------------------------------------------

    def tune_random_forest(self, X, y, dataset_key):

        self.config.log_execution(f"Tuning RF for {dataset_key}")

        param_grid = {
            "n_estimators": [50, 100, 200],
            "max_depth": [10, 20, None],
            "min_samples_split": [2, 5, 10]
        }

        rf = RandomForestClassifier(
            random_state=RANDOM_SEED,
            n_jobs=-1
        )

        grid_search = GridSearchCV(
            rf,
            param_grid,
            cv=3,
            scoring="f1_weighted",
            n_jobs=-1,
            verbose=0
        )

        if X.shape[0] > 5000:

            idx = np.random.choice(X.shape[0], 5000, replace=False)

            X_sample = X[idx]
            y_sample = y[idx]

        else:

            X_sample = X
            y_sample = y

        grid_search.fit(X_sample, y_sample)

        print("\nBest Parameters:", grid_search.best_params_)
        print("Best F1:", grid_search.best_score_)

        return grid_search.best_estimator_, grid_search.best_params_


# ============================================================================

tuner = HyperparameterTuner(CONFIG)

cv_results = []

cv_dir = CONFIG.dirs["reports"]
os.makedirs(cv_dir, exist_ok=True)


# ============================================================================
# CROSS VALIDATION
# ============================================================================

print("\n>>> Cross Validation <<<")

for dataset_key in ["anxiety", "depression"]:

    if dataset_key not in balanced_datasets:
        continue

    cv_file = f"{cv_dir}/{dataset_key}_cv.pkl"

    # ------------------------------------------------
    # SKIP IF ALREADY DONE
    # ------------------------------------------------

    if os.path.exists(cv_file):

        print(f"{dataset_key}: CV already done — loading")

        result = joblib.load(cv_file)

        cv_results.append(result)

        continue

    # ------------------------------------------------
    # RUN CV
    # ------------------------------------------------

    X = balanced_datasets[dataset_key]["X"]
    y = balanced_datasets[dataset_key]["y"]

    rf_model = RandomForestClassifier(
        n_estimators=100,
        random_state=RANDOM_SEED,
        n_jobs=-1
    )

    scores = tuner.perform_cross_validation(
        rf_model,
        X,
        y,
        dataset_key,
        "random_forest"
    )

    result = {
        "dataset": dataset_key,
        "model": "random_forest",
        "cv_mean": scores.mean(),
        "cv_std": scores.std()
    }

    joblib.dump(result, cv_file)

    cv_results.append(result)


# ============================================================================
# SAVE CV RESULTS TABLE
# ============================================================================

cv_df = pd.DataFrame(cv_results)

cv_table_path = f"{cv_dir}/cross_validation_results.csv"

cv_df.to_csv(cv_table_path, index=False)

CONFIG.log_execution(f"Saved CV table: {cv_table_path}")


# ============================================================================
# HYPERPARAMETER TUNING
# ============================================================================

print("\n>>> Hyperparameter Tuning <<<")

tuned_model_path = f"{CONFIG.dirs['models']}/anxiety_random_forest_tuned.pkl"

if os.path.exists(tuned_model_path):

    print("RandomForest already tuned — loading model")

    best_rf = joblib.load(tuned_model_path)

else:

    if "anxiety" in balanced_datasets:

        X = balanced_datasets["anxiety"]["X"]
        y = balanced_datasets["anxiety"]["y"]

        best_rf, best_params = tuner.tune_random_forest(X, y, "anxiety")

        joblib.dump(best_rf, tuned_model_path)

        CONFIG.log_execution(f"Saved tuned model: {tuned_model_path}")


print("\n" + "=" * 80)
print("STEP 9 COMPLETE: Cross-validation and tuning finished")
print("=" * 80 + "\n")
```

## PATTERN-BASED PREDICTION SYSTEM

```python
# ============================================================================

print("\n" + "=" * 80)
print("STEP 10: PATTERN-BASED PREDICTION SYSTEM")
print("=" * 80 + "\n")

import os
import pandas as pd

class PatternPredictor:

    def __init__(self, config):
        self.config = config
        self.pattern_rules = self._define_pattern_rules()

    # ------------------------------------------------------------------

    def _define_pattern_rules(self):

        return {
            'anxiety': {
                'keywords': ['worry', 'anxious', 'panic', 'fear', 'nervous', 'stress', 'overwhelm'],
                'threshold': 2
            },
            'depression': {
                'keywords': ['sad', 'hopeless', 'worthless', 'empty', 'alone', 'depressed', 'suicide'],
                'threshold': 2
            },
            'suicide_risk': {
                'keywords': ['suicide', 'kill', 'die', 'death', 'end', 'hurt', 'pain'],
                'threshold': 1
            }
        }

    # ------------------------------------------------------------------

    def predict_from_text(self, text):

        text_lower = text.lower()

        predictions = {}

        for condition, rule in self.pattern_rules.items():

            keyword_count = sum(
                1 for keyword in rule['keywords'] if keyword in text_lower
            )

            prediction = keyword_count >= rule['threshold']

            confidence = min(
                keyword_count / (rule['threshold'] * 2),
                1.0
            )

            predictions[condition] = {
                "prediction": prediction,
                "confidence": confidence,
                "keyword_count": keyword_count
            }

        return predictions

    # ------------------------------------------------------------------

    def batch_predict(self, texts):

        results = []

        for text in texts:

            pred = self.predict_from_text(text)

            record = {}

            for condition, values in pred.items():

                record[f"{condition}_pred"] = values["prediction"]
                record[f"{condition}_conf"] = values["confidence"]

            results.append(record)

        return pd.DataFrame(results)


# ============================================================================

pattern_predictor = PatternPredictor(CONFIG)

print("\n>>> Pattern-Based Predictions <<<")

# ---------------------------------------------------------------------------
# TEST EXAMPLES (ALWAYS RUN — VERY FAST)
# ---------------------------------------------------------------------------

test_texts = [
    "I feel so anxious and worried all the time, I can't stop panicking.",
    "Everything feels hopeless and I don't see any point in continuing.",
    "I'm having a great day, feeling happy and optimistic!",
    "The stress is overwhelming me, I'm constantly nervous and scared."
]

for i, text in enumerate(test_texts):

    print(f"\nText {i+1}: {text}")

    prediction = pattern_predictor.predict_from_text(text)

    for condition, result in prediction.items():

        if result["prediction"]:

            print(
                f"  {condition.upper()} detected "
                f"(confidence {result['confidence']:.2f})"
            )


# ---------------------------------------------------------------------------
# DATASET SAMPLE PREDICTIONS
# ---------------------------------------------------------------------------

pattern_pred_path = f"{CONFIG.dirs['reports']}/pattern_predictions_sample.csv"

if os.path.exists(pattern_pred_path):

    print("\nPattern predictions already computed — loading results")

    pattern_predictions = pd.read_csv(pattern_pred_path)

else:

    if "anxiety" in preprocessed_text_datasets:

        df_sample = preprocessed_text_datasets["anxiety"].sample(
            n=min(100, len(preprocessed_text_datasets["anxiety"]))
        )

        pattern_predictions = pattern_predictor.batch_predict(
            df_sample["processed_text"].tolist()
        )

        pattern_predictions.to_csv(pattern_pred_path, index=False)

        CONFIG.log_execution(
            f"Saved pattern predictions: {pattern_pred_path}"
        )


print("\nSample Prediction Statistics:")

print(pattern_predictions.describe())


print("\n" + "=" * 80)
print("STEP 10 COMPLETE: Pattern-based prediction system finished")
print("=" * 80 + "\n")
```

```python
# ============================================================================

print("\n" + "=" * 80)
print("STEP 13: FINAL VALIDATION AND OUTPUT VERIFICATION")
print("=" * 80 + "\n")

class OutputValidator:
    """
    Validates pipeline outputs and performs final checks.
    """

    def __init__(self, config: PipelineConfig):
        """
        Initialize OutputValidator.

        Args:
            config: Pipeline configuration object
        """
        self.config = config
        self.validation_results = {}

    def validate_datasets(self) -> bool:
        """
        Validate that all datasets were processed.

        Returns:
            True if validation passed
        """
        print(">>> Validating Datasets <<<")

        all_valid = True
        for dataset_key in self.config.datasets.keys():
            if dataset_key not in cleaned_datasets:
                print(f"[WARNING] Dataset {dataset_key} was not processed")
                all_valid = False
            else:
                df = cleaned_datasets[dataset_key]
                if df is None or len(df) == 0:
                    print(f"[WARNING] Dataset {dataset_key} is empty")
                    all_valid = False
                else:
                    print(f"[OK] Dataset {dataset_key}: {len(df)} samples")

        self.validation_results['datasets'] = all_valid
        return all_valid

    def validate_models(self) -> bool:
        """
        Validate that models were trained for each dataset.

        Returns:
            True if validation passed
        """
        print("\n>>> Validating Models <<<")

        all_valid = True
        for dataset_key in cleaned_datasets.keys():
            if dataset_key not in model_trainer.trained_models:
                print(f"[WARNING] No models trained for {dataset_key}")
                all_valid = False
            else:
                models = model_trainer.trained_models[dataset_key]
                if len(models) == 0:
                    print(f"[WARNING] No models found for {dataset_key}")
                    all_valid = False
                else:
                    print(f"[OK] Dataset {dataset_key}: {len(models)} models trained")

        self.validation_results['models'] = all_valid
        return all_valid

    def validate_outputs(self) -> bool:
        """
        Validate that output files were created.

        Returns:
            True if validation passed
        """
        print("\n>>> Validating Output Files <<<")

        required_files = [
            (f"{self.config.dirs['reports']}/model_performance.csv", "Model Performance CSV"),
            (f"{self.config.dirs['reports']}/pipeline_reports.json", "Pipeline Reports JSON"),
            (f"{self.config.dirs['reports']}/pipeline_report.html", "HTML Report")
        ]

        all_valid = True
        for filepath, description in required_files:
            if os.path.exists(filepath):
                file_size = os.path.getsize(filepath)
                print(f"[OK] {description}: {file_size} bytes")
            else:
                print(f"[WARNING] {description} not found: {filepath}")
                all_valid = False

        self.validation_results['outputs'] = all_valid
        return all_valid

    def count_artifacts(self) -> Dict[str, int]:
        """
        Count generated artifacts.

        Returns:
            Dictionary of artifact counts
        """
        print("\n>>> Counting Artifacts <<<")

        counts = {
            'plots': len([f for f in os.listdir(self.config.dirs['plots_eda']) if f.endswith('.png')]),
            'models': len([f for f in os.listdir(self.config.dirs['models']) if f.endswith('.pkl')]),
            'reports': len([f for f in os.listdir(self.config.dirs['reports'])]),
            'cached_items': len([f for f in os.listdir(self.config.dirs['cache']) if f.endswith('.pkl')])
        }

        print(f"Total EDA Plots: {counts['plots']}")
        print(f"Total Saved Models: {counts['models']}")
        print(f"Total Reports: {counts['reports']}")
        print(f"Total Cached Items: {counts['cached_items']}")

        return counts

    def generate_validation_report(self):
        """Generate final validation report."""
        print("\n" + "=" * 60)
        print("VALIDATION REPORT")
        print("=" * 60)

        for check, result in self.validation_results.items():
            status = "PASSED" if result else "FAILED"
            print(f"{check.upper()}: {status}")

        all_passed = all(self.validation_results.values())

        print("\n" + "=" * 60)
        if all_passed:
            print("PIPELINE VALIDATION: ALL CHECKS PASSED")
        else:
            print("PIPELINE VALIDATION: SOME CHECKS FAILED (see warnings above)")
        print("=" * 60)


# Initialize validator
validator = OutputValidator(CONFIG)

# Run validation checks
datasets_valid = validator.validate_datasets()
models_valid = validator.validate_models()
outputs_valid = validator.validate_outputs()

# Count artifacts
artifact_counts = validator.count_artifacts()

# Generate validation report
validator.generate_validation_report()

print("\n" + "=" * 80)
print("STEP 13 COMPLETE: Final validation finished")
print("=" * 80 + "\n")
```

```python
import os
import joblib
import onnx
from skl2onnx import convert_sklearn
from skl2onnx.common.data_types import FloatTensorType
import onnxmltools # Import onnxmltools to potentially register XGBoost converter

# =============================================================================
# ONNX EXPORT CONFIGURATION
# =============================================================================
ONNX_EXPORT_DIR = "/content/drive/MyDrive/Mental Health v1/exported_pipelines/onnx"
os.makedirs(ONNX_EXPORT_DIR, exist_ok=True)


class ONNXConverter:
    """
    Handles conversion of trained scikit-learn models to ONNX format.
    """

    def __init__(self, config, model_trainer):
        self.config = config
        self.model_trainer = model_trainer
        self.conversion_log = []

    def convert_sklearn_to_onnx(self, model, model_name, dataset_key, input_shape):
        try:
            self.config.log_execution(
                f"Converting {dataset_key} - {model_name} to ONNX"
            )

            # Define initial types for skl2onnx conversion
            initial_type = [('float_input', FloatTensorType([None, input_shape]))]

            # Convert scikit-learn model to ONNX
            onnx_model = convert_sklearn(model, initial_types=initial_type)

            onnx_path = f"{ONNX_EXPORT_DIR}/{dataset_key}_{model_name}.onnx"

            with open(onnx_path, "wb") as f:
                f.write(onnx_model.SerializeToString())

            self.conversion_log.append({
                "dataset": dataset_key,
                "model": model_name,
                "status": "success",
                "path": onnx_path
            })

            print(f"  Successfully converted and saved: {onnx_path}")
            return True

        except Exception as e:
            error_msg = f"Conversion failed for {dataset_key} - {model_name}: {str(e)}"
            print(f"  [ERROR] {error_msg}")
            self.conversion_log.append({
                "dataset": dataset_key,
                "model": model_name,
                "status": "failed",
                "reason": str(e)
            })
            return False

    def save_conversion_log(self):
        if not self.conversion_log:
            return

        df_log = pd.DataFrame(self.conversion_log)
        log_path = f"{self.config.dirs['reports']}/onnx_conversion_log.csv"
        df_log.to_csv(log_path, index=False)
        self.config.log_execution(f"Saved ONNX conversion log to: {log_path}")


# =============================================================================
# Perform ONNX Conversion
# =============================================================================

print("\n" + "=" * 80)
print("STEP: ONNX MODEL CONVERSION")
print("=" * 80 + "\n")

if 'model_trainer' not in globals() or 'all_metrics' not in globals():
    print("[ERROR] Required global objects (model_trainer, all_metrics) are not defined.")
    print("Please ensure that all preceding setup and model training cells have been executed successfully.")
else:
    onnx_converter = ONNXConverter(CONFIG, model_trainer)

    for dataset_key, models in model_trainer.trained_models.items():
        if dataset_key not in model_trainer.test_splits:
            print(f"[WARNING] Test splits not found for {dataset_key}, skipping ONNX conversion.")
            continue

        input_dim = model_trainer.test_splits[dataset_key]["X_test"].shape[1]

        dataset_metrics = [m for m in all_metrics if m["dataset"] == dataset_key]
        if not dataset_metrics:
            print(f"[WARNING] No metrics found for {dataset_key}, cannot determine best model for ONNX conversion.")
            continue

        # Filter out XGBoost models from consideration for ONNX conversion
        non_xgboost_metrics = [m for m in dataset_metrics if m["model"] not in ["xgboost", "lightgbm"]]

        # Sort remaining models by F1-score in descending order
        non_xgboost_metrics.sort(key=lambda x: x["f1_score"], reverse=True)

        if not non_xgboost_metrics:
            print(f"[WARNING] No compatible non-XGBoost/LightGBM models found for {dataset_key}. Skipping ONNX conversion for this dataset.")
            continue

        onnx_converted = False
        for metric in non_xgboost_metrics:
            model_to_convert_name = metric["model"]
            if model_to_convert_name not in models:
                print(f"[WARNING] Model '{model_to_convert_name}' not found in trained models for {dataset_key}. Skipping.")
                continue

            model = models[model_to_convert_name]
            print(f"Processing {dataset_key}: Attempting to convert model {model_to_convert_name} (F1: {metric['f1_score']:.4f})")

            if onnx_converter.convert_sklearn_to_onnx(
                model,
                model_to_convert_name,
                dataset_key,
                input_dim
            ):
                onnx_converted = True
                break # Successfully converted, move to next dataset

        if not onnx_converted:
            print(f"[WARNING] All compatible models failed ONNX conversion for {dataset_key}. No ONNX model generated for this dataset.")

    onnx_converter.save_conversion_log()

print("\n" + "=" * 80)
print("ONNX CONVERSION STEP COMPLETE")
print("=" * 80 + "\n")
```

```python
import gradio as gr
import onnxruntime as ort
import joblib
import os
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin
from scipy.sparse import hstack, csr_matrix

# =============================================================================
# Custom Class Definitions (Required to load the preprocessing components)
# =============================================================================

class TextFeatureBuilder(BaseEstimator, TransformerMixin):

    def __init__(self, vectorizer, scaler):
        self.vectorizer = vectorizer
        self.scaler = scaler

    def fit(self, X, y=None):
        return self

    def transform(self, texts):
        if isinstance(texts, str):
            texts = [texts]

        tfidf = self.vectorizer.transform(texts)

        numeric_features = []
        for text in texts:
            char_count = len(text)
            words = text.split()
            word_count = len(words)
            avg_word_length = np.mean([len(w) for w in words]) if words else 0
            sentence_count = max(text.count("."), 1)
            uppercase_ratio = (sum(1 for c in text if c.isupper()) / max(len(text), 1))

            numeric_features.append([
                char_count,
                word_count,
                avg_word_length,
                sentence_count,
                uppercase_ratio
            ])

        numeric_features = np.array(numeric_features, dtype="float32")
        numeric_scaled = self.scaler.transform(numeric_features)
        numeric_sparse = csr_matrix(numeric_scaled)

        return hstack([tfidf, numeric_sparse])

# =============================================================================
# Configuration
# =============================================================================
ONNX_EXPORT_DIR = "/content/drive/MyDrive/Mental Health v1/exported_pipelines/onnx"

# Label mappings for user-friendly output
label_mappings = {
    "anxiety": {0: "No Anxiety", 1: "Anxiety"}, # Note: Anxiety target was mapped to 1
    "depression": {0: "No Depression", 1: "Depression"}, # Note: Depression target was mapped to 1
    "suicide": {0: "No Suicide Risk", 1: "Suicide Risk"}, # Note: Suicide target was mapped to 1
    "stress": {0: "Low Stress", 1: "High Stress"},
    "bipolar": {0: "Non-Bipolar", 1: "Bipolar"}
}

# =============================================================================
# Load all ONNX models and their preprocessing components
# =============================================================================

onnx_models = {}
preprocessing_pipelines = {}

print("\nLoading ONNX models and preprocessing pipelines...")

for dataset_key in ["anxiety", "depression", "suicide", "stress", "bipolar"]:
    try:
        # Determine the model name for this dataset from the ONNX log
        # This assumes the ONNX log is available and updated correctly
        onnx_log_path = f"{CONFIG.dirs['reports']}/onnx_conversion_log.csv"
        if os.path.exists(onnx_log_path):
            onnx_log_df = pd.read_csv(onnx_log_path)
            dataset_log = onnx_log_df[(onnx_log_df['dataset'] == dataset_key) & (onnx_log_df['status'] == 'success')]
            if not dataset_log.empty:
                model_name = dataset_log.iloc[0]['model']
            else:
                print(f"[WARNING] No successful ONNX conversion found for {dataset_key} in log. Skipping.")
                continue
        else:
            print(f"[ERROR] ONNX conversion log not found at {onnx_log_path}. Cannot determine model names.")
            break

        onnx_model_path = os.path.join(ONNX_EXPORT_DIR, f"{dataset_key}_{model_name}.onnx")

        if not os.path.exists(onnx_model_path):
            print(f"[WARNING] ONNX model file not found for {dataset_key} at {onnx_model_path}. Skipping.")
            continue

        # Load ONNX session
        session = ort.InferenceSession(onnx_model_path)
        onnx_models[dataset_key] = session

        # Load Python preprocessing components
        tfidf_cache = CONFIG.load_from_cache(dataset_key, "tfidf_features")
        vectorizer = tfidf_cache["vectorizer"]

        selector_cache = CONFIG.load_from_cache(dataset_key, "chi2_features")
        selector = selector_cache["selector"]

        scaler = feature_engineer.scalers[f"{dataset_key}_text_scaler"]

        feature_builder = TextFeatureBuilder(vectorizer, scaler)

        preprocessing_pipelines[dataset_key] = {
            "feature_builder": feature_builder,
            "selector": selector
        }
        print(f"  Loaded {dataset_key} (Model: {model_name}) successfully.")

    except Exception as e:
        print(f"[ERROR] Failed to load ONNX model or preprocessing for {dataset_key}: {e}")

print("ONNX model loading complete.")

# =============================================================================
# Unified Prediction Logic for ONNX models
# =============================================================================
def predict_all_onnx(text_input):
    if not text_input or str(text_input).strip() == "":
        return {"Message": "Please enter text to analyze."}

    all_results = {}
    for dataset_key, onnx_session in onnx_models.items():
        try:
            pipeline_components = preprocessing_pipelines.get(dataset_key)
            if not pipeline_components:
                all_results[dataset_key.capitalize()] = "Preprocessing components not loaded."
                continue

            feature_builder = pipeline_components["feature_builder"]
            selector = pipeline_components["selector"]

            # 1. Preprocess the text using Python components
            features_raw = feature_builder.transform([text_input])

            # 2. Apply feature selection
            features_selected = selector.transform(features_raw)

            # 3. Convert to appropriate numpy array for ONNX runtime
            input_name = onnx_session.get_inputs()[0].name
            input_tensor = features_selected.astype(np.float32).toarray()

            # 4. Run inference
            outputs = onnx_session.run(None, {input_name: input_tensor})

            # Extract predicted class ID and probabilities
            predicted_class_id = outputs[0][0]
            probability_map = outputs[1][0]

            # Map numeric result to user-friendly label
            predicted_label = label_mappings[dataset_key].get(predicted_class_id, f"Unknown ({predicted_class_id})")

            all_results[dataset_key.capitalize()] = {
                "predicted_label": predicted_label,
                "class_probabilities": {str(k): float(v) for k, v in probability_map.items()} # Ensure JSON serializable
            }

        except Exception as e:
            all_results[dataset_key.capitalize()] = f"Error during ONNX prediction: {str(e)}"

    return all_results

# =============================================================================
# Gradio UI Construction
# =============================================================================
with gr.Blocks(theme=gr.themes.Soft()) as onnx_prediction_iface:
    gr.Markdown("# 🧠 Mental Health Analysis Pipeline (ONNX Version)")
    gr.Markdown("Enter text to get predictions for various mental health indicators using ONNX-exported models.")

    with gr.Row():
        with gr.Column():
            text_input_onnx_all = gr.Textbox(
                label="Text Input: Describe how you feel",
                placeholder="E.g., I feel overwhelmed and anxious, struggling with daily tasks.",
                lines=7
            )

            run_btn_onnx_all = gr.Button("Get ONNX Predictions", variant="primary")

        with gr.Column():
            gr.Markdown("### 📊 Prediction Results (ONNX Models)")
            output_json_onnx_all = gr.JSON(label="Model Predictions for Text Input (ONNX)")

    run_btn_onnx_all.click(
        fn=predict_all_onnx,
        inputs=[text_input_onnx_all],
        outputs=[output_json_onnx_all]
    )

onnx_prediction_iface.launch(share=True)
```
