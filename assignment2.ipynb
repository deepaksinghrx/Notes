```python
import pandas as pd
import numpy as np

try:
    simon = pd.read_csv("SIMON_pheno.csv")
except UnicodeDecodeError:
    simon = pd.read_csv("SIMON_pheno.csv", encoding="latin1")

print(simon.head())
```

```python
simon.shape
```

```python
simon.columns
```

```python
simon = simon.rename(columns={simon.columns[0]: "Session"})
print(simon.columns)
```

```python
simon.info()
```

```python
simon["Acquisition_date"] = pd.to_datetime(simon["Acquisition_date"], errors="coerce")
print(simon.dtypes)
```

```python
print(simon.isnull().sum())
```

```python
simon=simon.drop_duplicates()
```

## EDA

```python
simon.describe(include="all")
```

```python
cat_cols = simon.select_dtypes(include="object").columns
print("Categorical columns:", list(cat_cols))
```

```python
num_cols = simon.select_dtypes(include=[np.number]).columns
print("Numerical columns:", list(num_cols))
```

```python
print("Min Age:", simon["Age"].min())
print("Max Age:", simon["Age"].max())
print("Age range:", simon["Age"].max() - simon["Age"].min())
```

```python
print("Early acquisition:", simon["Acquisition_date"].min())
print("Latest acquisition:", simon["Acquisition_date"].max())
```

```python
if "manufacturer" in simon.columns and "Age" in simon.columns:
    print("\n=== Average Age by manufacturer ===")
    print(simon.groupby("manufacturer")["Age"].mean().sort_values())
```
