```python
try:
    with open(patients_data, 'r') as f:
        data = json.load(f)

    with open(finance_data, 'r') as f:
        finance_data = json.load(f)

    with open(health_data, 'r') as f:
        health_data = json.load(f)
except FileNotFoundError as e:
    print(f"File not found: {e}")
except Exception as e:
    print(f"An error occurred: {e}")

patient_df = pd.DataFrame(data['patients']) 
health_df = pd.DataFrame(health_data['healthRecords'])
finance_df = pd.DataFrame(finance_data['payments'])
```

```python
def data_report(df):
    print(df.head())
    print("Data Summary:")
    print(df.describe(include='all'))
    print("Data Types:")
    print(df.dtypes)
    print("\nMissing Values:")
    print(df.isna().sum())
    print("\nDuplicate Rows:")
    print(df.duplicated().sum())
```

```python
print("Patient Data Report:")
data_report(patient_df)
```

```python
patient_df['age']=patient_df['age'].fillna(patient_df['age'].mean())
patient_df['contact']=patient_df['contact'].fillna('not provided')
patient_df['gender']=patient_df['gender'].replace({'Male': 'M', 'Female': 'F'})
patient_df=patient_df.drop_duplicates()

print("Cleaned Patient Data Report:")
print(patient_df)
```

```python
cleaned_patient_df = patient_df.copy()
```

```python
cleaned_patient_df
```

```python
print("Patient Health Report:")
data_report(health_df)
```

```python
health_df['bloodGroup']=health_df['bloodGroup'].fillna('Unknown')
health_df['heightCm']=health_df['heightCm'].fillna(health_df['heightCm'].mean())
health_df['weightKg']=health_df['weightKg'].fillna(health_df['weightKg'].mean())
health_df=health_df.drop_duplicates()

print("Cleaned Health Data Report:")
print(health_df)
```

```python
cleaned_health_df = health_df.copy()
```

```python
print("Financial Data Report:")
data_report(finance_df)
```

```python
finance_df['paymentMode']=finance_df['paymentMode'].fillna('Cash')
finance_df['insurance']=finance_df['insurance'].replace({'Yes': 'Y', 'No': 'N'})
finance_df=finance_df.drop_duplicates()
print("Cleaned Financial Data Report:")
print(finance_df)   
```

```python
cleaned_finance_df = finance_df.copy()
```

```python
cleaned_patient_df.to_csv('cleaned_patients.csv', index=False)
cleaned_health_df.to_csv('cleaned_health.csv', index=False)
cleaned_finance_df.to_csv('cleaned_finance.csv', index=False)
```

```python
print("combined csv data")
combined_data=pd.merge(cleaned_patient_df, cleaned_health_df, on='patientId', how='inner')
combined_data=pd.merge(combined_data, cleaned_finance_df, on='patientId', how='inner')
combined_data.to_csv('combined_data.csv', index=False)
combined_data
```
